```python
class Tree(object):
  def __init__(self):
    self.left = None
    self.child = []
    self.data = []
  def createChildren(self,amount):
    for i in range(0,amount):
      self.child.append(Tree())
  def setChildrenValues(self,list):
    for i in range(0,len(list)):
      self.data.append(list[i])
  def setChildrenChildren(self, my_list, list2, list3, list4, list5, list6, list7):
    for i in range(0,len(my_list)):
      self.data.append([my_list[i], list2[i], list3[i], list4[i], list5[i], list6[i], list7[i]])
```

## Using a DCL (Dependency Constraint Language) to check structural dependencies

<p>
    <b>Process</b>
</p>
<p> 
1. Define Dependency Constraints -> 2. Analysis Source Code -> 3. List Architectural Violoations
</p>

<p>
<b>
Example of Dependency Constraints
</b>
</p>
<p>
    1: <b>only</b> Factory <b>can-create</b> DAO
</p>
<p>
    2: Util <b>can-depend-only</b> JavaAPI
</p>
<p>
    3: View <b>cannot-access</b> Model
</p>
<p>
4: DTO <b>must-implement</b> Serializable
</p>

<p>
<p>
These constraints state that only classes in the Factory module can create objects of classes in the DAO module (line 1); 
</p>
<p>
classes in module Util can establish dependencies only with classes from the Java API (line 2); 
</p>
<p>
classes in module View cannot access classes from module Model (line 3); 
</p>
<p>
every class in the DTO module must implement Serializable (line 4).    
</p>
</p>

<p>
    DCL allows developers to specify dependencies caused by accessing methods and fields (access), declaring variables (declare), creating objects (create), extending classes (extend), implementing interfaces (implement), throwing exceptions (throw), or using annotations (annotate). It is also possible to define constrains including any form of dependency (depend). (Terra, 2012)
</p>
